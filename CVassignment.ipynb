```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = "bob.jpg"  # Replace with your image path
img = cv2.imread(image_path)

# Verify that the image was loaded correctly
if img is None:
    print("Error: Image not found or could not be read. Please check the file path.")
else:
    # Low Pass Filter (Heavy blur)
    def low_pass_filter(img):
        kernel = np.ones((15, 15), np.float32) / 225  # Larger kernel for stronger blur
        return cv2.filter2D(img, -1, kernel)

    # Mean Filter (Moderate blur, preserving more edges)
    def mean_filter(img):
        return cv2.blur(img, (5, 5))  # Smaller kernel to retain some edges

    # Gaussian Filter (Smooth, high-quality blur)
    def gaussian_filter(img):
        return cv2.GaussianBlur(img, (7, 7), 1)  # Gaussian blur with moderate kernel

    # Apply the filters
    low_pass_img = low_pass_filter(img)
    mean_img = mean_filter(img)
    gaussian_img = gaussian_filter(img)

    # Convert images to RGB for display with matplotlib
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    low_pass_img_rgb = cv2.cvtColor(low_pass_img, cv2.COLOR_BGR2RGB)
    mean_img_rgb = cv2.cvtColor(mean_img, cv2.COLOR_BGR2RGB)
    gaussian_img_rgb = cv2.cvtColor(gaussian_img, cv2.COLOR_BGR2RGB)

    # Display the images
    plt.figure(figsize=(12, 8))

    plt.subplot(2, 2, 1)
    plt.imshow(img_rgb)
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(2, 2, 2)
    plt.imshow(low_pass_img_rgb)
    plt.title("Low Pass Filter")
    plt.axis("off")

    plt.subplot(2, 2, 3)
    plt.imshow(mean_img_rgb)
    plt.title("Mean Filter")
    plt.axis("off")

    plt.subplot(2, 2, 4)
    plt.imshow(gaussian_img_rgb)
    plt.title("Gaussian Filter")
    plt.axis("off")

    plt.tight_layout()
    plt.show()
```
